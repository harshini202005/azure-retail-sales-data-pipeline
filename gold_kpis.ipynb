```python
df = spark.read.format("delta").load(
    "abfss://silver@retaildatalakeproject.dfs.core.windows.net/clean_sales"
)
```

```python
from pyspark.sql.functions import *
```

```python
#KPI 1 — Total Revenue by Region
revenue_df = df.groupBy("Region").agg(
    round(sum("Sales"), 2).alias("Total_Revenue")
)
```

```python
#KPI 2 — Top Selling Products
top_products_df = df.groupBy("Product_Name").agg(
    round(sum("Sales"), 2).alias("Revenue")
).orderBy(desc("Revenue"))
```

```python
#KPI 3 — Monthly Sales Trend
monthly_sales_df = df.groupBy(
    "Sales_Year",
    "Sales_Month"
).agg(
    round(sum("Sales"), 2).alias("Monthly_Sales")
).orderBy("Sales_Year", "Sales_Month")
```

```python
#KPI 4 — Profit by Category
profit_df = df.groupBy("Category").agg(
    round(sum("Profit"), 2).alias("Total_Profit")
)
```

```python
#KPI 5 — Quantity Sold by Sub-Category
quantity_df = df.groupBy("Sub_Category").agg(
    sum("Quantity").alias("Total_Quantity")
)
```

```python
#KPI 6 — State Performance
state_df = df.groupBy("State").agg(
    round(sum("Sales"), 2).alias("Total_Sales"),
    round(sum("Profit"), 2).alias("Total_Profit")
)
```

```python
#KPI 7 — Customer Segment Analysis
segment_df = df.groupBy("Segment").agg(
    round(sum("Sales"), 2).alias("Segment_Sales")
)
```

```python
#Display KPIs
display(revenue_df)
display(top_products_df)
display(monthly_sales_df)
display(profit_df)
display(quantity_df)
display(state_df)
display(segment_df)
```

```python
#Save Gold Tables
#Revenue
revenue_df.write.format("delta") \
    .mode("overwrite") \
    .save(
        "abfss://gold@retaildatalakeproject.dfs.core.windows.net/revenue"
    )
```
